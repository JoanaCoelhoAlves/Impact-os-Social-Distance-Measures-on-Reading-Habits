For the second part of the analysis the dependent variable is built based on the results from equation 2 from the Sim et al model (2022) which i use in the first part of the analysis.

Stress, complaince and loneliness came from the COVIDiSTRESS survey, I only cleaned and scaled (the original dataset already display variables values at country level and the participants answer the survey with a peridiocity of 1 week).
So the key point is the dependent variable

model2 <- lm(log(total_books) ~ factor(country_code) + treated +  after + after_treated*factor(country_code) +

The dependent variable will be the value of after_treated*factor(country_code) for our reference country (Argentina); all the other countries i add the estimate of the main effect (after_treated*factor(country_code) for argentina) PLUS the respective country effect that arises from after_treated*factor(country_code)

```python
#this are the outputs from R which will be the values for the dependent variable after applying the requiered transformations: exp() and -1
values = [5.350e-01, -1.493e-01,-3.084e-01,-3.807e-01,-2.835e-01,-3.032e-01, -3.745e-01,-1.283e-01,3.727e-04,-1.534e-01,-8.698e-01,
          -3.758e-01,-2.075e-01,1.580e-02,-4.159e-02,-4.962e-01 ,-5.554e-01,-2.854e-01,-4.672e-01,3.330e-02,-1.123e-01,
          -7.911e-01, 1.849e-01,-1.955e-01,-7.046e-01,-3.753e-01,-2.325e-01,-1.935e-01,-2.937e-01,-4.443e-01,-8.158e-01,
          -4.219e-01,9.469e-02,-1.121e+00,-1.189e-01,-3.458e-01,3.198e-01,-4.008e-01, -5.527e-01,-4.006e-01,-5.493e-01,-2.630e+00,-3.677e-01]
```

```python
#check if there is a match in quantity of values and countries
len(values) == len(country_list)
```

```python
#objective is to do this for all values: np.exp(5.350e-01)-1
transformed_values =[]
for i in values:
    transformation=np.exp(i)-1
    transformed_values.append(transformation)

#make sure there is a match 
len(transformed_values)  == len(values) == len(country_list)  
    
```

```python
#create the data frame, do not request indexes in order to facilitate the next step where we marge dataframes
df2= pd.DataFrame({"country_code": country_list,
                  "y": transformed_values})
```

```python
df2["y_final"]=df2[["country_code","y"]].apply(lambda row: row[1] +0.707448 if row[0] != "ARG" else row[1], axis=1)
df2.head()
```

Join df2 and grouped_emotions together based on country_code.
Both dataframes have the desiered variables and both have 35 entries, so everything looks okay to exist a perfect match so we should a "inner" merge

```python
df2_final=pd.merge(df2,grouped_emotions,how="inner",on="country_code")
df2_final
```

Because in this research I'm not interest in the correlations between X,W,XW and because w (compliance) is a continous variable i need to use mena-centering technique to properly code my moderator (W).
mean_centering=Xi - Mean_X

So i need to calculate the mean of variable W (com)
Then i need to subtract for each row value the mean of W (com), that is my w_final 
Multiply x*W_final to get my correct and desiered value

```python
com_mean=df2_final["com"].mean()
com_mean
```

```python
df2_final["com_final"]=df2_final["com"]-com_mean
df2_final
```

```python
#rename variables to facilitate the R equations already built anddrop variable y since we only care about y_final
df_regression=df2_final.drop(columns=["y", "com"])
df_regression.head()
```

```python
df_regression.rename(columns= {"y_final":"y", "stress": "m1","lon":"x","com_final":"w"},inplace=True)
df_regression.head()
```

```python
#because i will want a first stage moderation i need to create a new variable xw which is the multiplication of x and w
df_regression["xw"]=df_regression["x"]*df_regression["w"]
df_regression.head()
```
